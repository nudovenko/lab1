```python
import pandas as pd
import math

def stint(x):
    return ord(x)/1000.0
def stint1(x):
    return 1.0 if x=='e' else 0.0

xlsx=pd.ExcelFile(r'C:\Users\Nastya\Documents\Udovenko.xlsx')

namescol=[]
convert={}

namescol = ['0','BB','CC','DD','EE','FF','GG','HH','II','JJ','KK','LL','MM','NN','OO','PP','QQ','RR','SS','TT','UU','VV','WW']

convert[namescol[0]]=stint1

for i in range(1,23):
    convert[namescol[i]]=stint

df=pd.read_excel(xlsx,usecols='A:W',header=None,names=namescol,nrows=50,converters=convert)
print(df)

#sorted_df=df.sort_values(by='CC')

#sorted_df.to_excel(r'C:\Users\Nastya\Documents\Udovenko2.xlsx',header=False, index=False)

#aa=df['CC'].value_counts()
#print(aa)

#with pd.ExcelWriter(r'C:\Users\Nastya\Documents\UdovenkoLab1.xlsx') as writer:
    #df.to_excel(writer)
    
elem_count = df.shape[0]
print('elem_count=',elem_count)

oo={}
pp={}
oo=df['0'].value_counts()
pp=df['0'].value_counts()/elem_count
Entr=-pp[0]*math.log(pp[0])-pp[1]*math.log(pp[1])
print(oo)
print('Entr=',Entr)

def calcEntr(columnName):
   ss={}
   ss=df[columnName].value_counts()
   k=[]
   for s in ss.keys():
    k.append(s)
     # print('keys = ',k)
    ngroup=len(k)
    el=[]
    for i in range(ngroup):
        el.append(ss[k[i]])
    dd=[]
    for i in range(ngroup):
        dd.append([])
    for j in range(2):
        dd[i].append(0.0)
    for i in range(elem_count):
        for j in range(ngroup):
                if df[columnName][i]==k[j]:
                    if df['0'][i]==0.0:
                        dd[j][0]=dd[j][0]+1.
                    else:
                        dd[j][1]=dd[j][1]+1.
    Entropies=[]
    for i in range(ngroup):
        EG=0.
        if dd[i][0]!=0.:
            EG=EG-dd[i][0]/el[i]*math.log(el[i][0]/el[i])
        if dd[i][1]!=0.:
            EG=EG-dd[i][1]/el[i]*math.log(el[i][1]/el[i])
        Entropies.append(EG)
    IG=Entr
    for i in range(ngroup):
        IG=IG-Entropies[i]*el[i]/elem_count

    return round(IG,3)
colNames = df.columns.tolist()
colNames = colNames[1:]
print()
print(colNames)

infGain = {}
for columnName in colNames:
    infGain[columnName] = calcEntr(columnName)

print()
print('information gain =', infGain)
```
